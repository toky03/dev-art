```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Anzahl_pixel_X=4000
Anzahl_pixel_Y=4000
Anfang_X=-0.836
Ende_X=-0.834
Anfang_Y=-0.233
Ende_Y=-0.231
```

```python
Tiefe=70
```

```python
Schrittlaenge_X=(Ende_X-Anfang_X)/Anzahl_pixel_X
```

```python
Schrittlaenge_Y=(Ende_Y-Anfang_Y)/Anzahl_pixel_Y
```

```python
x=np.linspace(Anfang_X,Ende_X,Schrittlaenge_X)
```

```python
x = np.arange(Anfang_X, Ende_X, Schrittlaenge_X)
```

```python
y = np.arange(Anfang_Y, Ende_Y, Schrittlaenge_Y)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
n = len(x)
```

```python
c = np.zeros([n,n])
```

```python
z0 = np.add(X, np.multiply(Y, 1j))
```

```python
z = np.zeros([n,n])
```

```python
for i in range(Tiefe):
    z=np.power(z,2)+z0
    c[np.abs(z) < 2] = i
```

```python
plt.imshow(c, cmap='viridis')
```
